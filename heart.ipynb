```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data= pd.read_csv('heart.csv')
data
```

```python
data['sex'].value_counts()
```

```python
labels = ['Male', 'Female']
plt.pie(data['sex'].value_counts(), labels=labels, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
data['age'].describe()
```

```python
plt.hist(data['age'], bins= 10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
bins = [0, 18, 35, 50, 65, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Middle-aged', 'Senior']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)
print(data['age_group'].value_counts())
```

```python
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 
           'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
corr_matrix = data[columns].corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True )
plt.title("Correlation Matrix of HEart Disease")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data, x='trestbps', color= 'green')
plt.title("Box Plot of BP")
plt.xlabel("TrestBPS")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['thalach'], y=data['chol'], hue=data['target'], palette={0: 'blue', 1: 'red'})
plt.xlabel('Max Heart Rate (thalach)')
plt.ylabel('Cholesterol (chol)')
plt.title('Max Heart Rate vs Cholesterol')
plt.legend(title="Heart Disease")  
plt.show()
```

```python
pd.crosstab(data.target, data.sex).plot.bar(color=['red', 'blue'])
plt.title('Heart Disease by Gender')
plt.xlabel('Heart Disease Status (0 = No Disease, 1 = Disease)')
plt.ylabel('Count')
plt.legend(["Female", "Male"])
plt.xticks(rotation=0)
plt.show()
```
